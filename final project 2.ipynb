```python

fp = open('exp.txt', 'r')
line = fp.readline()
s = line.split()
no_data = 1 #有興趣資料所在位子
data = []
while line:
    s = line.split()
    if len(s) < (no_data+1) :
        line = fp.readline()
    else :
        data.append(float(s[1]))
        line = fp.readline()
fp.close()

import matplotlib.pyplot as plt
import numpy as np
l = len(data) #數據總數
x = np.linspace(0,l, l)
plt.plot(x, data,'.')
plt.show()

ac=10 #分割找數據段數
dataa = []
xa = []
for i in range (ac) :
    dataa.append(data[int(l/ac*i):int(l/ac*(i+1))])
    xa.append(x[int(l/ac*i):int(l/ac*(i+1))])
    plt.plot(xa[i],dataa[i])
```
