```python
fp = open('4.txt', 'r')
line = fp.readline()
s = line.split()
no_data = 1 #
data = []
while line:
    s = line.split()
    if len(s) < (no_data+1) :
        line = fp.readline()
    else :
        data.append(1000*float(s[1]))
        line = fp.readline()
fp.close()

import matplotlib.pyplot as plt
import numpy as np
l = len(data) #
x = np.linspace(0,l, l)
#plt.plot(x, data,'.')

from scipy import optimize

def func(x, a, b, c):
    return a*np.exp(-b*x)+c
x=np.linspace(1,l,num=l)


ac=10 #
dataa = []
xa = []
for i in range (ac) :
    dataa.append(data[int(l/ac*i):int(l/ac*(i+1))])
    xa.append(x[int(l/ac*i):int(l/ac*(i+1))])
    plt.plot(xa[i],dataa[i], '.')
    popt, pcov = optimize.curve_fit(func, xa, dataa, maxfev=500000)
    a = popt[0] 
    b = popt[1]
    c = popt[2]
    yvals = func(x,a,b,c)
    print (u'a:', a)
    print (u'b:', b)
    print (u'c:', c)
    plt.plot(x, yvals)
    plt.show()
```
